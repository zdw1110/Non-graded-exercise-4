```python
import random
import time

# Store all cards in a list of dictionaries
card_deck = [
    {"name": "Diablo", "health": 100, "attack": 90, "defense": 60},
    {"name": "Medusa", "health": 100, "attack": 70, "defense": 70},
    {"name": "Jester", "health": 120, "attack": 60, "defense": 90},
    {"name": "Troll", "health": 150, "attack": 40, "defense": 94},
    {"name": "Specter", "health": 100, "attack": 70, "defense": 70},
    {"name": "Mist", "health": 100, "attack": 75, "defense": 65},
    {"name": "Savage", "health": 100, "attack": 90, "defense": 50},
    {"name": "Marauder", "health": 100, "attack": 85, "defense": 50},
    {"name": "Wimp", "health": 110, "attack": 40, "defense": 85},
    {"name": "Sorcerer", "health": 100, "attack": 70, "defense": 55}
]

# Set deck size
deck_size = 3

# Draw random cards for player and opponent
player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def display_card(card):
    """Display card details."""
    return f"Name: {card['name']}, Health: {card['health']}, Attack: {card['attack']}, Defense: {card['defense']}"

def card_battle(player_card, opponent_card):
    """Simulates a card battle between the player's card and the opponent's card."""
    print(f"\nYour card: {display_card(player_card)}")
    print(f"Opponent's card: {display_card(opponent_card)}")

    while player_card['health'] > 0 and opponent_card['health'] > 0:
        # Player's attack turn
        if player_card['attack'] >= opponent_card['health']:
            print(f"Your {player_card['name']} defeats the opponent's {opponent_card['name']}!")
            return "Player"
        else:
            # Attack the opponent
            damage = max(0, player_card['attack'] - opponent_card['defense'])
            opponent_card['defense'] = max(0, opponent_card['defense'] - player_card['attack'])
            if damage > 0:
                opponent_card['health'] -= damage
            print(f"After your attack: Opponent's {opponent_card['name']} Health: {opponent_card['health']}, Defense: {opponent_card['defense']}")
        
        # Check if opponent is defeated
        if opponent_card['health'] <= 0:
            print(f"Your {player_card['name']} defeats the opponent's {opponent_card['name']}!")
            return "Player"

        # Opponent's attack turn
        if opponent_card['attack'] >= player_card['health']:
            print(f"Opponent's {opponent_card['name']} defeats your {player_card['name']}!")
            return "Opponent"
        else:
            # Attack the player
            damage = max(0, opponent_card['attack'] - player_card['defense'])
            player_card['defense'] = max(0, player_card['defense'] - opponent_card['attack'])
            if damage > 0:
                player_card['health'] -= damage
            print(f"After opponent's attack: Your {player_card['name']} Health: {player_card['health']}, Defense: {player_card['defense']}")

        # Check if player is defeated
        if player_card['health'] <= 0:
            print(f"Opponent's {opponent_card['name']} defeats your {player_card['name']}!")
            return "Opponent"

        # Pause for the next round
        print("\nNext round in 3 seconds...")
        time.sleep(3)

while player_deck and opponent_deck:
    print("\nYour Deck:")
    for i, card in enumerate(player_deck):
        print(f"{i + 1}. {display_card(card)}")

    print("\nOpponent is choosing a card...")
    time.sleep(2)
    opponent_card = random.choice(opponent_deck)
    print(f"Opponent has chosen: {display_card(opponent_card)}")

    # Validate player input
    while True:
        try:
            player_choice = int(input(f"Choose your card to battle (1-{len(player_deck)}): ")) - 1
            if 0 <= player_choice < len(player_deck):
                break
            else:
                print("Invalid choice. Please choose a number within the range of your deck.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    player_card = player_deck[player_choice]

    winner = card_battle(player_card, opponent_card)

    if winner == "Player":
        opponent_deck.remove(opponent_card)
    else:
        player_deck.remove(player_card)

    if not opponent_deck:
        print("\nCongratulations! You have defeated all of the opponent's cards and won the game!")
        break
    elif not player_deck:
        print("\nYou have no more cards left. The opponent has won the game!")
        break

    print("\nA new round will begin in 3 seconds...")
    time.sleep(3)
```
